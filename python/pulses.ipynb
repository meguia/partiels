```python
import numpy as np
from scipy import signal as sg
import matplotlib.pyplot as plt
```

```python
pdur = 5.0
fs = 96000
nrep = 10
N = int(np.floor(fs*pdur))
npulses = np.zeros(N*nrep)
fc = 5000
```

```python
sos = sg.iirfilter(3, [fc], btype='lowpass',ftype='butter',fs=fs,output="sos") # diseno del filtro
w,h = sg.sosfreqz(sos,fs=fs)
dg = len(h)
pulse = np.zeros(2*dg)
pulse[dg-1] = 1.0
pulse[dg] = -1.0
pulsef = sg.sosfilt(sos,pulse) 
plt.plot(pulsef)
```

```python
for n in range(nrep):
    npulses[n*N:n*N+2*dg] = pulsef
plt.plot(npulses)    
```
